```python
from pubchem_search.browser import PubChemSearch

browser = PubChemSearch()

hits = browser.lookup_matching_smiles(
    smarts="[C;!$([c]);!$(C1CCCC=C1)]=[C;!$([c]);!$(C1CCCC=C1)]-[C;!$([c]);!$(C1CCCC=C1)]=[C;!$([c]);!$(C1CCCC=C1)]",
    limit=100,
    oversample=100,
    randomize=True,
    kind="ConnectivitySMILES",
)
hits
```
